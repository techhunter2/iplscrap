```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path="C:/Users/ReneRuiz/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
driver.get("https://www.iplt20.com/stats/2022/")
```

```python
vall_btn=driver.find_element_by_link_text("View All")
vall_btn.click()
```

```python
data=driver.page_source
data1=bs(data,"html.parser")
```

```python
msn=data1.find('table',class_="np-mostruns_table statsTable ng-scope")
```

```python
# title=[]
# for i in msn.find_all("th"):
#     title.append(i.text)
```

```python
pos=[]
pl=[]
match=[]
inn=[]
no=[]
runs=[]
hs=[]
avg=[]
bf=[]
sr=[]
hndrd=[]
fif=[]
four=[]
six=[]
rows= msn.find_all('tr')
```

```python
for row in rows:
    mn=row.find_all(class_='ng-binding')
    for i in range(0,14):
        for da in range(i,len(mn)):
            if da==0:
                pos.append(mn[da].text)
            elif da==1:
                pl.append(mn[da].text)
            elif da==2:
                match.append(mn[da].text)
            elif da==3:
                inn.append(mn[da].text)
            elif da==4:
                no.append(mn[da].text)
            elif da==5:
                runs.append(mn[da].text)
            elif da==6:
                hs.append(mn[da].text)
            elif da==7:
                avg.append(mn[da].text)
            elif da==8:
                bf.append(mn[da].text)
            elif da==9:
                sr.append(mn[da].text)
            elif da==10:
                hndrd.append(mn[da].text)
            elif da==11:
                fif.append(mn[da].text)
            elif da==12:
                four.append(mn[da].text)
            elif da==13:
                six.append(mn[da].text)
            break;
```

```python
df= {'POS':pos,'PLAYER':pl,'MATCH':match,'INNINGS':inn,'NO':no,'RUNS':runs,'HS':hs,'AVG':avg,'BF':bf,'SR':sr,'HUNDRED':hndrd,'FIFTY':fif,'FOUR':four,'SIX':six}
```

```python
new = pd.DataFrame.from_dict(df)
```
